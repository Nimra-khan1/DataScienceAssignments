```python
import pandas as pd

# Load dataset
df = pd.read_csv("house_price_cleaned.csv")

# Display first 5 rows
df.head()
```

```python
# Mean, Median, Mode, Variance
mean_price = df['price'].mean()
median_price = df['price'].median()
mode_price = df['price'].mode()[0]
variance_price = df['price'].var()

print("Mean Price:", mean_price)
print("Median Price:", median_price)
print("Mode Price:", mode_price)
print("Variance of Price:", variance_price)
```

```python
# Correlation matrix
correlation = df.corr(numeric_only=True)

# Sort correlation with 'price'
correlation_with_price = correlation['price'].sort_values(ascending=False)
print("Correlation of each feature with Price:\n")
print(correlation_with_price)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

  ###  Insights:
- The top 3 features most related to price are:
  1. sqft_living
  2. bathrooms
  3. bedrooms

These features have the highest positive correlation with price, indicating that as they increase, house prices also tend to rise.
