```python
import pandas as pd

data = {
    'SalePrice': [200000, 150000, 250000, 180000, 300000, 220000, 400000, 120000, 350000, 280000],
    'GrLivArea': [1400, 1200, 1600, 1300, 1800, 1500, 2000, 1100, 1700, 1550],
    'HouseStyle': ['1Story', '1Story', '2Story', '1Story', '2Story', '2Story', '2Story', '1Story', '2Story', '2Story'],
    'GarageCars': [2, 1, 2, 1, 3, 2, 3, 1, 2, 2],
    'OverallQual': [6, 5, 7, 6, 8, 6, 9, 5, 8, 7],
    'YearBuilt': [2005, 1995, 2010, 2000, 2015, 2008, 2020, 1990, 2012, 2007]
}

df = pd.DataFrame(data)
df

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 1️⃣ Histogram
plt.figure(figsize=(8,5))
sns.histplot(df['SalePrice'], kde=True)
plt.title("Distribution of House Prices")
plt.show()
# Insight:
print("Insight 1: The distribution of house prices is slightly right-skewed, meaning most houses are moderately priced while a few are very expensive.")
```

```python
# ---------- 2️⃣ Scatter Plot ----------
plt.figure(figsize=(8,5))
sns.scatterplot(x='GrLivArea', y='SalePrice', data=df, color='orange', s=90)
plt.title("Living Area vs Sale Price")
plt.show()
print("Insight 2: Larger houses generally have higher prices, showing a positive relationship between area and price.")
```

```python
# ---------- 3️⃣ Bar Plot ----------
plt.figure(figsize=(8,5))
sns.barplot(x='HouseStyle', y='SalePrice', data=df, palette='pastel')
plt.title("Average Price by House Style")
plt.show()
print("Insight 3: Two-story houses have higher average prices than one-story houses, indicating style impacts value.")
```

```python
# ---------- 4️⃣ Heatmap ----------
plt.figure(figsize=(10,6))
corr = df.select_dtypes(include=['number']).corr()
sns.heatmap(corr, cmap='coolwarm', annot=True, fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
print("Insight 4: SalePrice is highly correlated with OverallQual, GrLivArea, and GarageCars, meaning these affect price most.")
```

```python
# ---------- 5️⃣ Line Plot ----------
plt.figure(figsize=(8,5))
sns.lineplot(x='YearBuilt', y='SalePrice', data=df, marker='o', color='green')
plt.title("Year Built vs Sale Price")
plt.show()
print("Insight 5: Newer houses have higher prices, showing modern construction adds to property value.")
```
